```python
import matplotlib.pyplot as plt 
```

```python
def cost_function(x):
# given function
    return(x+3)**2
```

```python
def gradient(x):
    # derivative of given function
    return 2* (x+3)
```

```python
learning_rate=0.1
initial_x =2.0
num_itreation = 100
x_values = []
y_values = []
x = initial_x
for i in range(num_itreation):
    x_values.append(x)
    y_values.append(cost_function(x))
    gradient_value = gradient(x)
    x= x -learning_rate * gradient_value
    print(f'iteration {i+1}: x={x}, cost = {cost_function(x)}')
print(f'Optimal x:{x}')
```

```python
plt.plot(x_values,y_values,'ro-')
plt.title('gradient descent visulization for y =(x+3)^2 by AB')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
