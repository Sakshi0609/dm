```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
from sklearn.preprocessing import StandardScaler
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r"C:\Users\DELL\Downloads\lab 5\sales_data_sample.csv", encoding="latin")
```

```python
data = data[['ORDERLINENUMBER', 'SALES']]
```

```python
scaler = StandardScaler()
scaled_values = scaler.fit_transform(data.values)
```

```python
wcss = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, init='k-means++')
    model.fit_predict(scaled_values)
    wcss.append(model.inertia_)
```

```python
plt.plot(range(1, 11), wcss, 'ro-')
plt.show()
```

```python
model = KMeans(n_clusters=7, init='k-means++')
clusters = model.fit_predict(scaled_values)
clusters
```

```python
data['cluster'] = clusters
```
