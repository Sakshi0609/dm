```python

def solve_knapsack():
    val = [50, 100, 150, 200]  # Value array
    wt = [8, 16, 32, 40]  # Weight array
    W = 64  # Maximum weight capacity
    n = len(val)  # Number of items

    # Memoization table to store intermediate results
    memo = [[-1] * (W + 1) for _ in range(n)]

    # Helper function for recursive knapsack with memoization
    def knapsack(W, n):
        # Base case
        if n == 0 or W == 0:
            return 0

        # If value is already computed
        if memo[n - 1][W] != -1:
            return memo[n - 1][W]

        # If weight of current item is greater than remaining capacity, skip it
        if wt[n - 1] > W:
            memo[n - 1][W] = knapsack(W, n - 1)
        else:
            # Include the item and exclude the item, take max of both
            include_item = val[n - 1] + knapsack(W - wt[n - 1], n - 1)
            exclude_item = knapsack(W, n - 1)
            memo[n - 1][W] = max(include_item, exclude_item)
        
        return memo[n - 1][W]

    # Function to find out which items to include in the knapsack for optimal value
    def find_items(W, n):
        chosen_items = []
        remaining_weight = W

        for i in range(n, 0, -1):
            # If the current value does not equal the value of excluding the item, include it
            if memo[i - 1][remaining_weight] != memo[i - 2][remaining_weight]:
                chosen_items.append(i - 1)  # Append item index
                remaining_weight -= wt[i - 1]  # Reduce remaining weight

        return chosen_items

    # Calculate the maximum value and print it
    max_value = knapsack(W, n)
    print("Maximum value in Knapsack =", max_value)

    # Find the chosen items based on the computed memo table
    chosen_items = find_items(W, n)
    print("Items chosen (0-based index):", chosen_items)
    print("Items chosen (weights):", [wt[i] for i in chosen_items])
    print("Items chosen (values):", [val[i] for i in chosen_items])

# Run the function
solve_knapsack()


# In[2]:


def solve_knapsack():
    val=[50,100,150,200] #value array
    wt=[8,16,32,40] # Weight array
    W=64
    n=len(val) - 1
    def knapsack(W,n): # (Remaining Weight, Number of items checked)
        #base case
        if n<0 or W<=0:
            return 0
        
        #Higher weight than available
        if wt[n]>W:
            return knapsack(W, n-1)
        
        else:
            return max(val[n] + knapsack(W-wt[n],n-1),knapsack(W,n-1))
            # max(including , not including)
    print(knapsack(W,n))

solve_knapsack()
```
